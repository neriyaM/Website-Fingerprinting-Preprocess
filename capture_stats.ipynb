```python
def stats_sessions(labeled_sessions):
    lengths = []
    for s in labeled_sessions:
        lengths.append(len(s.session))
    print("Length: ", len(lengths))   
    print("Average: ", statistics.mean(lengths))
    print("Variance: ", statistics.pvariance(lengths))
    print("Max: ", max(lengths))
    print("Min: ", min(lengths))
```

```python
labels = {
    "facebook.com": 0,
    "google.com": 1,
    "instagram.com": 2,
    "jd.com": 3,
    "microsoft.com": 4,
    "sohu.com": 5,
    "wikipedia.org": 6,
    "yahoo.com": 7,
    "youtube.com": 8
}
labels
```

```python
def prepare_data(mainpage_handshakes):
    train_data = numpy.zeros((len(mainpage_handshakes), 1000))
    train_labels = numpy.zeros((len(mainpage_handshakes), 1))
    for idx, handshake in enumerate(mainpage_handshakes):
        data = numpy.array(handshake.session)
        data.resize(1000)
        train_data[idx] = data 
        train_labels[idx] = labels[handshake.label]
    return train_data, train_labels
```

```python
mainpage_train, mainpage_label = prepare_data(mainpage_handshakes)
```

```python
print('Training data shape : ', mainpage_train.shape, mainpage_label.shape)
```

```python
# Find the unique numbers from the train labels
classes = numpy.unique(mainpage_label)
nClasses = len(classes)
print('Total number of outputs : ', nClasses)
print('Output classes : ', classes)
```

```python
# Change to float datatype
train_data = mainpage_train.astype('float32')
# Scale the data to lie between 0 to 1
train_data /= 255
train_labels_one_hot = to_categorical(mainpage_label)
train_data = train_data.reshape(train_data.shape[0],train_data.shape[1], 1)
```

```python
def createModel():
    model = Sequential()
    # The first two layers with 32 filters of window size 3
    model.add(Conv1D(32, 3, padding='same', activation='relu', input_shape=(1000, 1)))
    model.add(Conv1D(32, 3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.25))

    model.add(Conv1D(64, 3, padding='same', activation='relu'))
    model.add(Conv1D(64, 3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.25))

    model.add(Conv1D(64, 3, padding='same', activation='relu'))
    model.add(Conv1D(64, 3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(nClasses, activation='softmax'))
    
    return model
```

```python
test_extractor = LabeledCapturesExtractor("C:\\Users\\Neriya\\Documents\\VM\\Shared\\test")
```

```python
test_captures = test_extractor.extract()
```

```python
test_handshakes = extract_mainpage_handshake(test_captures)
test_data, test_labels = prepare_data(test_handshakes)
```

```python
# Change to float datatype
test_data = test_data.astype('float32')
# Scale the data to lie between 0 to 1
test_data /= 255
test_labels_one_hot = to_categorical(test_labels)
test_data = test_data.reshape(test_data.shape[0],test_data.shape[1], 1)
```

```python
model = createModel()
batch_size = 256
epochs = 50
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(train_data, train_labels_one_hot, batch_size=batch_size, epochs=epochs, verbose=1, validation_data=(test_data, test_labels_one_hot))
```

```python
model.evaluate(test_data, test_labels_one_hot)
```
